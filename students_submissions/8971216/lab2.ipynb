#### First we'd need some sample data to plot for our graphs. ####

```python
import numpy as np
import random

# set a seed to get reporduceable results
np.random.seed(42)

# generate vectors of normally distributed values for plotting
x = np.random.normal(0, 1.0, size= (500,1))
y = np.random.normal(0, 1.0, size=(500, 1))
```

## Scatterplot ##

```python
import matplotlib.pyplot as plt

# the very basic scatterplot
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

Fig 1. A basic scatterplot

```python
# I can plot multiple plots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=False)

# I can also annotate the plot right in matplotlib 
# there are a couple of different coordinate systems I can use

# figure pixels is when (0,0) is the lower left hand corner
# and the xy argument defines the amount of pixels to offset
ax1.annotate('bottom left',
            xy=(10, 10), xycoords='figure pixels')

# figure fraction is where (0,0) is the lower left hand 
# corner, and (1,1) is the upper left hand corner
ax1.annotate('top left', \
            xy=(0.1, 0.9), xycoords= "figure fraction")

# I can mess with the sizes and colours 
# of the points in the scatterplot!
ax1.scatter(x,y, c='red', s=10)

# I can also add another plot into the same plot
ax2.scatter(x,y, c='green',s=5)

plt.show()
```

Fig 2. 2 plots side by side

## Histogram ##

Since we generated the random data using a normal distribution, it might be interesting to look at a histogram of the data.

```python
fig, ax = plt.subplots()
ax.hist(x)

plt.show()
```

Fig 3. A basic histogram

The histogram looks normally distributed!

I can introduce gridlines and outline the bars to make reading off values a little easier.

```python
fig, ax = plt.subplots()
ax.hist(x, edgecolor= "black")
plt.grid(visible= True)

plt.show()
```

Fig 4. Histogram with gridlines.

## 2D Histogram ##

We can try a fun one- 2D histogram. (Code taken from matplotlib's gallery)

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# we can see how changing the bins would quite dramatically change
# how the plot is interpreted. 
ax1.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)))
ax2.hist2d(x, y)

plt.show()
```

Fig 5. Two 2D histograms with different binning.
